```python
#importing packages needed for analysis
import pandas as pd
import numpy as np
import os

subfolders = [ f.path for f in os.scandir('inputs/') if f.is_dir() ]
print(subfolders)
```

```python
def read_df(subfolder,filestart,file_ends,tab_num,skiprow,start_year,end_year):
    eiaall = pd.DataFrame()
    temp = pd.DataFrame()
    print(eiaall.shape)
    print(temp.shape)
    for year in range (start_year,end_year+1):
        print(year)
        xlsx = pd.ExcelFile('inputs/'+subfolder+str(year)+filestart+str(year)+file_ends)
        temp = pd.read_excel(xlsx, tab_num, skiprows = skiprow)
        temp['YR'] = year
        print(temp.shape)
        eiaall = pd.concat([eiaall,temp], ignore_index=True)
        print(eiaall.shape)
        print(eiaall.columns[0:3])
    return eiaall
```

```python
#read in 860 data
subfolder = 'eia860'
filestart = '/3_1_Generator_Y'
file_ends = '.xlsx'
tab_num = 0
skiprow = 1
start_year = 2014
end_year = 2021

eia860 = read_df(subfolder,filestart,file_ends,tab_num,skiprow,start_year,end_year)
```
